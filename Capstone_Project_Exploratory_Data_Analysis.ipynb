## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Make a copy of the DataFrame
df_copy = df.copy()

# Drop rows with missing values in the 'name' column
df_copy.dropna(subset=['name'], inplace=True)

# Drop columns with more than 50% missing values
threshold = 0.5 * len(df_copy)
df_copy.dropna(thresh=threshold, axis=1, inplace=True)

# Fill missing values in numerical columns with median and categorical columns with mode
for column in df_copy.select_dtypes(include=['float64', 'int64']).columns:
    df_copy[column].fillna(df_copy[column].median(), inplace=True)

for column in df_copy.select_dtypes(include=['object']).columns:
    df_copy[column].fillna(df_copy[column].mode()[0], inplace=True)

# Cap outliers in the 'price' column at the 99th percentile
price_cap = df_copy['price'].quantile(0.99)
df_copy['price'] = df_copy['price'].apply(lambda x: price_cap if x > price_cap else x)

# Convert 'last_review' to datetime
df_copy['last_review'] = pd.to_datetime(df_copy['last_review'], errors='coerce')

# Create new features: Extract year and month from 'last_review'
df_copy['last_review_year'] = df_copy['last_review'].dt.year
df_copy['last_review_month'] = df_copy['last_review'].dt.month

# Final dataset inspection
print("\nFinal Dataset Info:")
print(df_copy.info())
print("\nFinal Dataset Description:")
print(df_copy.describe())

# Define the cleaned file path
cleaned_file_path = '/content/drive/MyDrive/Almabetter/EDA-Capstone-Project/Cleaned_Airbnb_NYC_2019.csv'

# Ensure the directory exists
os.makedirs(os.path.dirname(cleaned_file_path), exist_ok=True)

# Save the cleaned dataset
df_copy.to_csv(cleaned_file_path, index=False)
````

### What all manipulations have you done and insights you found?

A copy of the original DataFrame was made to preserve the integrity of the raw data during manipulations. Rows with missing values in the 'name' column were dropped to maintain data integrity, as the 'name' field is crucial for identification and analysis. Columns with more than 50% missing values were removed to enhance dataset quality, ensuring that analyses are not significantly impacted by incomplete data.

Numerical columns were filled with the median value of each respective column, as the median is less sensitive to outliers compared to the mean. Categorical columns were filled with the mode, or most frequent value, ensuring that the most common category is used to fill missing valuesa reasonable approach when specific information is lacking.

Outliers in the 'price' column were capped at the 99th percentile, preventing extreme values from skewing the analysis and providing a more robust representation of typical prices. The 'last_review' column was converted to a datetime format, facilitating date-related analysis and operations. Additionally, two new columns were created: 'last_review_year' and 'last_review_month', which extract the year and month from the 'last_review' column. This allows for temporal analysis, such as identifying trends over different periods.

These analyses can derive actionable insights that help stakeholders make informed decisions regarding their listings, pricing strategies, and marketing initiatives.
